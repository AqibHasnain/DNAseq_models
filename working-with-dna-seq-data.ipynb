#### 3 approaches for sequence to vector embedding. 1) encode the sequence information as an ordinal vector and work with that directly, 2) one-hot encode the sequence letters and use the resulting array and 3) treat the DNA sequence as a language (text) and use various "language" processing methods.

#### Ordinal encoding

```python
# A,T,C,G becomes 0.25,0.5,0.75,1.0, respectively.
# if there is an "N" because the base is uncertain, we will assign it to 0.0
def str_to_arr(seq):
    '''convert DNA sequence to numpy array'''
    seq = seq.lower()
    seq = re.sub('[^acgt]', 'z', seq)
    arr = np.array(list(seq))
    return arr

# create a label encoder with 'acgtn' alphabetx
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder();
label_encoder.fit(np.array(['a','c','g','t','z']));

def ordinal_encoder(arr):
    ''' function to encode a DNA sequence string as an ordinal vector.
     returns a numpy array with a=0.25, c=0.50, g=0.75, t=1.00, n=0.00'''
    integer_encoded = label_encoder.transform(arr)
    float_encoded = integer_encoded.astype(float)
    float_encoded[float_encoded == 0] = 0.25 # A
    float_encoded[float_encoded == 1] = 0.50 # C
    float_encoded[float_encoded == 2] = 0.75 # G
    float_encoded[float_encoded == 3] = 1.00 # T
    float_encoded[float_encoded == 4] = 0.00 # anything else, z
    return float_encoded

ordinal_encoder(str_to_arr(example_seq))
```

#### One-hot encoding of DNA sequence

```python
from sklearn.preprocessing import OneHotEncoder
def one_hot_encoder(arr):
    '''return an Nx4 matrix where N is the length of the sequence.'''
    integer_encoded = label_encoder.transform(arr)
    onehot_encoder = OneHotEncoder(sparse=False,dtype=int,categories=[range(4)])
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
#     onehot_encoded = np.delete(onehot_encoded, -1, 1)
    return onehot_encoded

one_hot_encoder(str_to_arr(example_seq))
```

#### Treating DNA sequence as a "language", otherwise known as k-mer counting
Both one-hot encoding and ordinal vectors are not ideal for classification or regression because sequences of differing lengths will produce a matrix or vector of varying length (can pad with zeros or truncate and still have it be useful) 

The sequence language analogy: genome as the book, subsequences (genes and gene families) are sentences and chapters, k-mers and peptides (motifs) are words, and nucleotide bases and amino acids are the alphabet. Since the analogy seems so apt, it stands to reason that the amazing work done in the natural language processing field should also apply to the natural language of DNA and protein sequences.

First take the long biological sequence and break it down into k-mer length overlapping “words”. For example, if the "words" are of length 6 (hexamers), “ATGCATGCA” becomes: ‘ATGCAT’, ‘TGCATG’, ‘GCATGC’, ‘CATGCA’. Hence our example sequence is broken down into 4 hexamer words. This "word" length is a hyperparameter that can be tuned

In genomics, these types of manipulations aka "k-mer counting", or counting the occurences of each possible k-mer sequence. 

```python
def getKmers(sequence, size):
    '''convert sequence to overlapping k-mer "words" '''
    return [sequence[x:x+size].lower() for x in range(len(sequence) - size + 1)]
getKmers(example_seq,6)
```

```python
# the "words" can be joined into a "sentence" and fed into NLP methods. 
words = getKmers(example_seq, size=6)
sentence = ' '.join(words)
sentence
# if the words are hexamers (length 6) and the alphabet has 4 elements, then there are 4^6 = 4096 possible words of length 6
```

```python
mySeq2 = 'GATGGCCATCCCCGCCCGAGCGGGGGGGG'
mySeq3 = 'CATGGCCATCCCCGCCCGAGCGGGCGGGG'
sentence2 = ' '.join(getKmers(mySeq2, size=6))
sentence3 = ' '.join(getKmers(mySeq3, size=6))
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform([sentence, sentence2, sentence3]).toarray() # counts of hexamers in the vocabulary
```
